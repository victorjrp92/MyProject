```python
# I filter the data to only include non-lethal attacks:
attacks = attacks[attacks["Fatal (Y/N)"] == "N"]
```

```python
# I filter the data for each country (the United States, Australia, and South Africa):
usa_attacks = attacks[attacks["Country"] == "USA"]
aus_attacks = attacks[attacks["Country"] == "AUSTRALIA"]
saf_attacks = attacks[attacks["Country"] == "SOUTH AFRICA"]
```

```python

# I take the data for each country based on the gender of the victim:
usa_women_attacks = usa_attacks[usa_attacks["Sex"] == "F"]
usa_men_attacks = usa_attacks[usa_attacks["Sex"] == "M"]

aus_women_attacks = aus_attacks[aus_attacks["Sex"] == "F"]
aus_men_attacks = aus_attacks[aus_attacks["Sex"] == "M"]

saf_women_attacks = saf_attacks[saf_attacks["Sex"] == "F"]
saf_men_attacks = saf_attacks[saf_attacks["Sex"] == "M"]
```

```python
#I count the number of non-lethal attacks for each country and gender:
usa_women_count = usa_women_attacks.shape[0]
usa_men_count = usa_men_attacks.shape[0]

aus_women_count = aus_women_attacks.shape[0]
aus_men_count = aus_men_attacks.shape[0]

saf_women_count = saf_women_attacks.shape[0]
saf_men_count = saf_men_attacks.shape[0]
```

```python
# Now I visualize the results with bar plots for each country:

import matplotlib.pyplot as plt

# plot the results for the USA
plt.bar(["Women", "Men"], [usa_women_count, usa_men_count])
plt.title("USA Non-Lethal Shark Attacks by Gender")
plt.xlabel("Gender")
plt.ylabel("Number of Attacks")
plt.show()

# plot the results for Australia
plt.bar(["Women", "Men"], [aus_women_count, aus_men_count])
plt.title("Australia Non-Lethal Shark Attacks by Gender")
plt.xlabel("Gender")
plt.ylabel("Number of Attacks")
plt.show()

# plot the results for South Africa
plt.bar(["Women", "Men"], [saf_women_count, saf_men_count])
plt.title("South Africa Non-Lethal Shark Attacks by Gender")
plt.xlabel("Gender")
plt.ylabel("Number of Attacks")
plt.show()
```
